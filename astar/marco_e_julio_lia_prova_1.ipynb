Definição da Heuristica

```python
import csv
csvh = open("data/heuristics_prova.csv")
grid = list(csv.reader(csvh))
heuristics = dict()
for i, row in enumerate(grid):
    heuristics[chr(i+97)] = dict()
    for j, col in enumerate(row):
        #if(col!='0.'):
        heuristics[chr(i+97)][chr(j+97)] = float(col)
print(f"H:\n\t{heuristics}")
```
